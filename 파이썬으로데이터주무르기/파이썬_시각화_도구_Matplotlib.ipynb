Numpy를 이용해서 sin을 만들고 그려보자

`arange` 명령으로 0부터 12까지 0.01 간격으로 데이터를 생성해서 `sin`으로 Sin값을 만든다.  
`t`라는 시간 혹은 그래프 상에서 x축을 의미하는 데이터를 0부터 12까지 만들고 그냥 사인 함수에 입력해서 그 출력을 `y`에 저장했다. 

```python
import numpy as np
t = np.arange(0, 12, 0.01)
y = np.sin(t)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, y)
plt.show()
```

`plot`명령에 `lw` 옵션으로 선 굵기를 설정하고, `color` 옵션으로 선 색깔을 설정한다.

```python
plt.figure(figsize=(10, 6))
plt.plot(t, np.sin(t), lw=3, label='sin')
plt.plot(t, np.cos(t), color='r', label='cos')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, y, color='green', linestyle='dashed')
plt.show()
```
