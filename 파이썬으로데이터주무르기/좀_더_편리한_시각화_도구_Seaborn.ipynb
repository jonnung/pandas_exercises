```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
import numpy as np
```

```python
x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2 * np.sin(x + 0.5)
y3 = 3 * np.sin(x + 1.0)
y4 = 4 * np.sin(x + 1.5)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

Seaborn은 연습할 만한 데이터셋을 몇개 가지고 있다.

```python
# 요일별 점심, 저녁, 흡연 여부와 식사 금액, 팁을 정리한 데이터
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, height=7)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, palette="Set1", height=7)
plt.show()
```

```python
# 연도 및 월별 항공기 승객수 데이터
flights = sns.load_dataset("flights")
flights.head(5)
```

```python
flights_pivot = flights.pivot("month", "year", "passengers")
flights_pivot.head(5)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d")
plt.show()
```

```python
# 머신러닝에서 중요하게 다뤄지는 아이리스 꽃에 대한 데이터
sns.set(style="ticks")
iris = sns.load_dataset("iris")
iris.head(10)
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```
